```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./res/tiempos_variando_alpha_k1_50_1_100.csv', sep=',', engine='python')
# print(df)

cant_nodos = df["accuracy"]
df["alpha"] = np.around(df["alpha"], decimals=0)
ax = sns.heatmap(data = df.groupby(['k','alpha'], as_index=False).mean().pivot('k','alpha', 'accuracy'),cmap="PuRd")
ax.set_title("Acuracy segun alpha y k".format(df["accuracy"]))
ax.invert_yaxis()
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)
plt.xlabel(r'alpha')
plt.ylabel(r'k')
plt.savefig('img/accuracy-alpha-k.png')
plt.show()
```
