```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./tiempos_variando_alpha_k1_200_1_100.csv', sep=',', engine='python')
# print(df)

cant_nodos = df["accuracy"]
df["alpha"] = np.around(df["alpha"], decimals=0)
ax = sns.heatmap(data = df.groupby(['k','alpha'], as_index=False).mean().pivot('k','alpha', 'accuracy'),cmap="PuRd")
ax.set_title("Accuracy variando alpha y k".format(df["accuracy"]))
ax.invert_yaxis()
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)
plt.xlabel(r'alpha')
plt.ylabel(r'k')
plt.savefig('img/accuracy-variando-alpha-k.png')
plt.show()
```

```python
df1 = pd.read_csv('./tiempos_variando_alpha_k1_200_1_100.csv', sep=',', engine='python')
df2 = []
for x in range(15, 186):
    df = df1.loc[df1['alpha'] == x] #agarro cada alpha
    mejork = 8
    mejoracc = 0
    for index, row in df.iterrows():
        if row['accuracy'] > mejoracc:
            mejoracc = row['accuracy']
            mejork = row['k']
    if mejoracc != 0:
        df2.append((x, mejork))

               
x_val = [x[0] for x in df2]
y_val = [x[1] for x in df2]

print (y_val)
plt.title('Mejor k para cada alpha')
plt.ylabel(r'k')
plt.xlabel(r'alpha')
plt.plot(x_val,y_val)
plt.plot(x_val,y_val,'or')
        
    
plt.savefig('./img/mejoresk-segunalpha.png')
plt.show()
```
