```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import decimal

```

```python
df = pd.read_csv('./tiempos_variando_alpha_k1_200_1_100.csv', sep=',', engine='python')
# print(df)

cant_nodos = df["accuracy"]
df["alpha"] = np.around(df["alpha"], decimals=0)
ax = sns.heatmap(data = df.groupby(['k','alpha'], as_index=False).mean().pivot('k','alpha', 'accuracy'),cmap="PuRd")
ax.set_title("Accuracy variando alpha y k".format(df["accuracy"]))
ax.invert_yaxis()
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)
plt.xlabel(r'alpha')
plt.ylabel(r'k')
plt.savefig('img/accuracy-variando-alpha-k.png', bbox_inches = "tight")
plt.show()
```

```python
df1 = pd.read_csv('./tiempos_variando_alpha_k1_200_1_100.csv', sep=',', engine='python')
df2 = []
for x in range(15, 186):
    df = df1.loc[df1['alpha'] == x] #agarro cada alpha
    mejork = 8
    mejoracc = 0
    tiempo = 0
    for index, row in df.iterrows():
        if row['accuracy'] > mejoracc:
            mejoracc = row['accuracy']
            mejork = row['k']
            tiempo = row['tiempo']
    if mejoracc != 0:
        df2.append((x, mejork, mejoracc, tiempo))
        
dff1 = pd.read_csv('./res_variando_alpha_k_inverted1_200_1_100.csv', sep=',', engine='python')
dff2 = []
for x in range(15, 186):
    dff = dff1.loc[dff1['alpha'].astype(int) == x] #agarro cada alpha pero no son enteros je
    mejork = 8
    mejoracc = 0
    tiempo = 0
    for index, row in dff.iterrows():
        if row['accuracy'] > mejoracc:
            mejoracc = row['accuracy']
            mejork = row['k']
            tiempo = row['tiempo']
    if mejoracc != 0:
        dff2.append((x, mejork, mejoracc, tiempo))
        
x_val = [x[0] for x in df2]
y_val = [x[2] for x in df2]
x_val_2 = [x[0] for x in dff2]
y_val_2 = [x[2] for x in dff2]

plt.title('Accuracy con alpha y k_alpha')
plt.ylabel(r'accuracy')
plt.xlabel(r'alpha')
plt.plot(x_val,y_val, 'g', label = "knn + pca")
plt.plot(x_val,y_val,'og')
plt.plot(x_val_2,y_val_2, 'r', label = "inverted knn + pca")
plt.plot(x_val_2,y_val_2,'or')
plt.legend()        
    
plt.savefig('./img/comparacion-inverted-mejork.png')
plt.show()
        
x_val = [x[0] for x in df2]
y_val = [x[3] for x in df2]
x_val_2 = [x[0] for x in dff2]
y_val_2 = [x[3] for x in dff2]

plt.title('Tiempo con alpha y k_alpha')
plt.ylabel(r'tiempo')
plt.xlabel(r'alpha')
plt.plot(x_val,y_val, 'g', label="knn + pca")
plt.plot(x_val,y_val,'og')
plt.plot(x_val_2,y_val_2, 'r',label="inverted knn + oca")
plt.plot(x_val_2,y_val_2,'or')
plt.legend()        
    
plt.savefig('./img/comparacion-inverted-mejork-tiempo.png')
plt.show()
```
