```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('res_pca_n_vs_sin_pca_1_50.csv', engine='python')

ax = sns.pointplot(x="k", y="accuracy", hue="metodo", data=df)
ax.set_xlabel("k")
ax.set_ylabel("accuracy")
ax.set_title("Comp. accuracy variando k con PCA=#vocabs vs sin-PCA")
ax.grid()

plt.savefig('img/accuracy-pca-n-vs-sin-pca.png')
```

```python
df = pd.read_csv('res_pca_n_vs_sin_pca_1_50.csv', engine='python')

ax = sns.pointplot(x="k", y="tiempo", hue="metodo", data=df)
ax.set_xlabel("k")
ax.set_ylabel("tiempo (ms)")
ax.set_title("Comp. tiempos ms variando k con PCA=#vocabs vs sin-PCA")
ax.grid()

plt.savefig('img/tiempos-pca-n-vs-sin-pca.png')
```
