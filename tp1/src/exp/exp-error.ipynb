```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
#import plotly.graph_objs as go
```

```python
df = pd.read_csv('resultados/random-var-m-p-data.csv', sep=',', engine='python')
# print(df)

cant_nodos = df["n"][0]
df["p"] = np.around(df["p"], decimals=2)
ax = sns.heatmap(data = df.groupby(['p','p_densidad'], as_index=False).mean().pivot('p','p_densidad', 'val'),cmap="Blues")
ax.set_title("Error de cuentas según densidades y p".format(cant_nodos))
ax.invert_yaxis()
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)
plt.xlabel(r'Páginas')
plt.ylabel(r'p')
plt.savefig('img/random-var-m-p.png')
plt.show()

df = pd.read_csv('resultados/random-var-m-p-data-mucha-densidad.csv', sep=',', engine='python')
# print(df)

cant_nodos = df["n"][0]
df["p"] = np.around(df["p"], decimals=2)
ax = sns.heatmap(data = df.groupby(['p','p_densidad'], as_index=False).mean().pivot('p','p_densidad', 'val'),cmap="Blues")
ax.set_title("Error de cuentas según densidades y p".format(cant_nodos))
ax.invert_yaxis()
ax.set_yticklabels(ax.get_yticklabels(), rotation = 0)
plt.xlabel(r'Páginas')
plt.ylabel(r'p')
plt.savefig('img/random-var-m-p.png')
plt.show()
```
