```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('resultados/random-var-m-data.csv', sep=',', engine='python')
# print(df)
cant_nodos = df["n"][0]
print(cant_nodos)
cant_pruebas = df["cant_intentos"][0]
df["p_densidad"] = np.around(df["p_densidad"], decimals=2)
ax = sns.pointplot(x="p_densidad", y="tiempo_ns", data=df)
plt.xlabel(r'Densidad')
plt.ylabel(r'Tiempo (ns)')
print("Cantidad de pruebas para valor: {}".format(cant_pruebas))
# ax.set_title("Tiempos variando #links; #págs={}".format(cant_nodos))

for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 10 == 1:
        label.set_visible(True)
    else:
        label.set_visible(False)
        
plt.savefig('img/tiempos-random-var-m.png')
plt.show()
```

```python
df = pd.read_csv('resultados/random-var-m-data.csv', sep=',', engine='python')
# print(df)

n = df["n"][0]
m_completo = n*(n-1)
aristas = df['p_densidad'].apply(lambda x: int(m_completo * x))
aristas = aristas.apply(lambda x: x)
sns.jointplot(aristas, df['tiempo_ns'], kind="reg", x_estimator=np.mean)
plt.xlabel(r'Densidad')
plt.ylabel(r'Tiempo (ns)')
plt.savefig('img/tiempos-random-var-m-rlineal.png')
plt.show()
```
