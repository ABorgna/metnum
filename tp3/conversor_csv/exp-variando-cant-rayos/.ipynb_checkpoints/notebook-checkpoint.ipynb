```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import seaborn as sns
```

```python
img = mpimg.imread('../../data/tomo-torso.png')
plt.imshow(img)
plt.show()

df = pd.read_csv('exp_res.csv', engine='python')

df = df[df["nombre_img"] == "tomo-torso"] 
# print(df)


# df['max_vocab_freq'] = df['max_vocab_freq'].apply(lambda x: round(x,2))

ax = sns.pointplot(x="ray_cnt", y="tiempo_lsq", hue="ray_type", data=df)
# ax.set_xlabel("Max. Vocabs. Freq.")
# ax.set_ylabel("Tiempo (ms)")
# ax.set_title("Tiempos variando frecuencia mx. en vocabulario")
ax.grid()
plt.locator_params(nbins=10)

# plt.savefig('img/tiempos-variando-max-vocabs-freq.png', bbox_inches = "tight")
plt.show()


ax = sns.pointplot(x="ray_cnt", y="psnr", hue="ray_type", data=df)
# ax.set_xlabel("Max. Vocabs. Freq.")
# ax.set_ylabel("Tiempo (ms)")
# ax.set_title("Tiempos variando frecuencia mx. en vocabulario")
ax.grid()
plt.locator_params(nbins=10)

# plt.savefig('img/tiempos-variando-max-vocabs-freq.png', bbox_inches = "tight")

plt.show()
```
