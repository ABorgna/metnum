```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('resultados/trivial-var-p-data.csv', sep=',', engine='python')

cant_nodos = df["n"][0]
df["p"] = np.around(df["p"], decimals=2)

ax = sns.heatmap(data = df.groupby(['p','nodo'], as_index=False).mean().pivot('p','nodo', 'val_nodo'))
ax.set_title("Resultados de grafo trivial de {} nodos variando p".format(cant_nodos))
ax.invert_yaxis()
plt.savefig('img/trivial-var-p.png')
plt.show()
```

```python
df = pd.read_csv('resultados/completo-var-p-data.csv', sep=',', engine='python')

cant_nodos = df["n"][0]
df["p"] = np.around(df["p"], decimals=2)

ax = sns.heatmap(data = df.groupby(['p','nodo'], as_index=False).mean().pivot('p','nodo', 'val_nodo'))
ax.set_title("Resultados de grafo completo de {} nodos variando p".format(cant_nodos))
ax.invert_yaxis()
plt.savefig('img/completo-var-p.png')
plt.show()
```

```python
df = pd.read_csv('resultados/union-estrellas-in-var-p-data.csv', sep=',', engine='python')

cant_estrellas = df["cant_estrellas"][0]
cant_nodos_por_estrella = df["cant_nodos_por_estrella"][0]
df["p"] = np.around(df["p"], decimals=2)

ax = sns.heatmap(data = df.groupby(['p','nodo'], as_index=False).mean().pivot('p','nodo', 'val_nodo'), cmap="Blues")
ax.set_title("Resultados de grafo union hacia adentro de estrellas ({},{}) variando p".format(cant_estrellas, cant_nodos_por_estrella))
ax.invert_yaxis()
plt.savefig('img/union-estrellas-in.png')
plt.show()
```

```python
df = pd.read_csv('resultados/union-estrellas-out-var-p-data.csv', sep=',', engine='python')

cant_estrellas = df["cant_estrellas"][0]
cant_nodos_por_estrella = df["cant_nodos_por_estrella"][0]
df["p"] = np.around(df["p"], decimals=2)

ax = sns.heatmap(data = df.groupby(['p','nodo'], as_index=False).mean().pivot('p','nodo', 'val_nodo'),cmap="Blues")
ax.set_title("Resultados de grafo union hacia afuera de estrellas ({},{}) variando p".format(cant_estrellas, cant_nodos_por_estrella))
ax.invert_yaxis()
plt.savefig('img/union-estrellas-out.png')
plt.show()
```
