```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('resultados/random-var-m-data.csv', sep=',', engine='python')
# print(df)
cant_nodos = df["n"][0]
df["p_densidad"] = np.around(df["p_densidad"], decimals=2)
ax = sns.pointplot(x="p_densidad", y="tiempo_ns", data=df)
ax.set_title("Tiempo de ejecuci√≥n grafos random de {} nodos variando m".format(cant_nodos))
ax.grid()

for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 2 == 1:
        label.set_visible(True)
    else:
        label.set_visible(False)
        
plt.savefig('img/random-var-m.png')
plt.show()
```
