```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('resultados/random-var-n-m-fijo-data.csv', sep=',', engine='python')


df['cant_nodos'] = df['cant_nodos'].apply(lambda x: x**3)
# print(df)
sns.jointplot(df['cant_nodos'], df['tiempo_ns'], kind="reg")
# plt.xlabel(r'${2^n}n$')
# plt.ylabel(r'Tiempo (ns)')
# plt.savefig('../../../../pics/tiempos-exacto-variando-n-mcompleto.png')
plt.show()

df2 = pd.read_csv('resultados/random-var-n-m-ajustado-data.csv', sep=',', engine='python')


df2['cant_nodos'] = df2['cant_nodos'].apply(lambda x: x**3)
# print(df)
sns.jointplot(df2['cant_nodos'], df2['tiempo_ns'], kind="reg")
# plt.xlabel(r'${2^n}n$')
# plt.ylabel(r'Tiempo (ns)')
# plt.savefig('../../../../pics/tiempos-exacto-variando-n-mcompleto.png')
plt.show()
```

```python
df = pd.read_csv('resultados/random-var-n-m-fijo-data.csv', sep=',', engine='python')

# print(df)
ax = sns.pointplot(x="cant_nodos", y="tiempo_ns", data=df)
# ax.set_title("Tiempo de ejecución grafos random de {} nodos variando m".format(cant_nodos))
# ax.grid()
        
for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 3 == 1:
        label.set_visible(True)
    else:
        label.set_visible(False)
        
    
plt.savefig('img/random-var-m.png')
plt.show()


df2 = pd.read_csv('resultados/random-var-n-m-ajustado-data.csv', sep=',', engine='python')

# print(df)
ax = sns.pointplot(x="cant_nodos", y="tiempo_ns", data=df2)
# ax.set_title("Tiempo de ejecución grafos random de {} nodos variando m".format(cant_nodos))
# ax.grid()
        
for ind, label in enumerate(ax.get_xticklabels()):
    if ind % 4 == 1:
        label.set_visible(True)
    else:
        label.set_visible(False)
        
    
plt.savefig('img/random-var-m.png')
plt.show()
```
